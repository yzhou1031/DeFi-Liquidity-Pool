```python
import requests

RPC = "https://api.mainnet-beta.solana.com"
address = "4KPZc461qPHF1bKcGn8ecu6dJoofAi3QMnQXkqH82g2S"  # or a token account address

payload = {
    "jsonrpc": "2.0",
    "id": 1,
    "method": "getSignaturesForAddress",
    "params": [
        address,
        {"limit": 50}  # up to 1000, but start small
    ]
}

res = requests.post(RPC, json=payload).json()
if "error" in res:
    raise RuntimeError(res["error"])

sigs = res["result"]
print("num signatures:", len(sigs))
print(sigs[0])  # shows signature + blockTime + slot
```
